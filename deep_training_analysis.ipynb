# DEEP Model - Training Analysis

DetaylÄ± training analizi ve grafikler

```
# Setup
from google.colab import drive, files
drive.mount('/content/drive')

import json
import matplotlib.pyplot as plt
import numpy as np
import glob

print("âœ“ KÃ¼tÃ¼phaneler yÃ¼klendi")
```

```
# Log'larÄ± yÃ¼kle
log_dir = '/content/drive/MyDrive/lora_checkpoints/deep/logs'
log_files = glob.glob(f"{log_dir}/training_log_*.jsonl")

if not log_files:
    print("âŒ Log dosyasÄ± bulunamadÄ±!")
else:
    log_file = log_files[0]
    print(f"âœ“ Log dosyasÄ±: {log_file}")
    
    with open(log_file, 'r') as f:
        logs = [json.loads(line) for line in f]
    
    # AyÄ±r
    train_logs = [log for log in logs if 'train_loss' in log]
    eval_logs = [log for log in logs if 'eval_loss' in log]
    
    print(f"âœ“ Train logs: {len(train_logs)}")
    print(f"âœ“ Eval logs: {len(eval_logs)}")
```

```
# Ä°statistikler
train_steps = [log['step'] for log in train_logs]
train_losses = [log['train_loss'] for log in train_logs]
learning_rates = [log.get('learning_rate', 0) for log in train_logs]

print("ðŸ“Š DEEP Model Ä°statistikleri")
print("=" * 50)
print(f"Toplam Step: {len(train_logs)}")
print(f"Ä°lk Train Loss: {train_losses[0]:.4f}")
print(f"Son Train Loss: {train_losses[-1]:.4f}")
print(f"Ä°yileÅŸme: {((train_losses[0] - train_losses[-1]) / train_losses[0] * 100):.1f}%")
print(f"Min Loss: {min(train_losses):.4f}")
print(f"Max Loss: {max(train_losses):.4f}")

if eval_logs:
    eval_losses = [log['eval_loss'] for log in eval_logs]
    print(f"\nEn Ä°yi Eval Loss: {min(eval_losses):.4f}")
    best_idx = eval_losses.index(min(eval_losses))
    print(f"En Ä°yi Checkpoint: step-{eval_logs[best_idx]['step']}")
```

```
# Grafik 1: Training Loss
plt.figure(figsize=(12, 6))
plt.plot(train_steps, train_losses, 'b-', linewidth=2, alpha=0.7, label='Train Loss')
plt.xlabel('Step', fontsize=14, fontweight='bold')
plt.ylabel('Loss', fontsize=14, fontweight='bold')
plt.title('DEEP Model - Training Loss Over Time', fontsize=16, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('deep_train_loss.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
print("âœ“ Grafik 1 kaydedildi: deep_train_loss.png")
```

```
# Grafik 2: Validation Loss
if eval_logs:
    eval_steps = [log['step'] for log in eval_logs]
    eval_losses = [log['eval_loss'] for log in eval_logs]
    
    plt.figure(figsize=(12, 6))
    plt.plot(eval_steps, eval_losses, 'r-', linewidth=2.5, marker='o', markersize=8, label='Validation Loss')
    plt.xlabel('Step', fontsize=14, fontweight='bold')
    plt.ylabel('Loss', fontsize=14, fontweight='bold')
    plt.title('DEEP Model - Validation Loss', fontsize=16, fontweight='bold')
    plt.grid(True, alpha=0.3, linestyle='--')
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.savefig('deep_eval_loss.png', dpi=300, bbox_inches='tight', facecolor='white')
    plt.show()
    print("âœ“ Grafik 2 kaydedildi: deep_eval_loss.png")
```

```
# Grafik 3: Learning Rate Schedule
if any(learning_rates):
    plt.figure(figsize=(12, 6))
    plt.plot(train_steps, learning_rates, 'g-', linewidth=2, label='Learning Rate')
    plt.xlabel('Step', fontsize=14, fontweight='bold')
    plt.ylabel('Learning Rate', fontsize=14, fontweight='bold')
    plt.title('DEEP Model - Learning Rate Schedule', fontsize=16, fontweight='bold')
    plt.grid(True, alpha=0.3, linestyle='--')
    plt.legend(fontsize=12)
    plt.ticklabel_format(style='scientific', axis='y', scilimits=(0,0))
    plt.tight_layout()
    plt.savefig('deep_learning_rate.png', dpi=300, bbox_inches='tight', facecolor='white')
    plt.show()
    print("âœ“ Grafik 3 kaydedildi: deep_learning_rate.png")
```

```
# Grafik 4: Combined View
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Train Loss
axes[0, 0].plot(train_steps, train_losses, 'b-', linewidth=2)
axes[0, 0].set_xlabel('Step', fontweight='bold')
axes[0, 0].set_ylabel('Loss', fontweight='bold')
axes[0, 0].set_title('Training Loss', fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)

# Eval Loss
if eval_logs:
    axes[0, 1].plot(eval_steps, eval_losses, 'r-', linewidth=2, marker='o')
    axes[0, 1].set_xlabel('Step', fontweight='bold')
    axes[0, 1].set_ylabel('Loss', fontweight='bold')
    axes[0, 1].set_title('Validation Loss', fontweight='bold')
    axes[0, 1].grid(True, alpha=0.3)

# Learning Rate
if any(learning_rates):
    axes[1, 0].plot(train_steps, learning_rates, 'g-', linewidth=2)
    axes[1, 0].set_xlabel('Step', fontweight='bold')
    axes[1, 0].set_ylabel('Learning Rate', fontweight='bold')
    axes[1, 0].set_title('Learning Rate Schedule', fontweight='bold')
    axes[1, 0].grid(True, alpha=0.3)
    axes[1, 0].ticklabel_format(style='scientific', axis='y', scilimits=(0,0))

# Loss Improvement
window = min(10, len(train_losses))
if window > 1:
    smoothed = np.convolve(train_losses, np.ones(window)/window, mode='valid')
    smoothed_steps = train_steps[window-1:]
    axes[1, 1].plot(smoothed_steps[:len(smoothed)], smoothed, 'purple', linewidth=2)
else:
    axes[1, 1].plot(train_steps, train_losses, 'purple', linewidth=2)
axes[1, 1].set_xlabel('Step', fontweight='bold')
axes[1, 1].set_ylabel('Loss (Smoothed)', fontweight='bold')
axes[1, 1].set_title('Smoothed Training Loss', fontweight='bold')
axes[1, 1].grid(True, alpha=0.3)

plt.suptitle('DEEP Model - Complete Training Analysis', fontsize=18, fontweight='bold', y=1.00)
plt.tight_layout()
plt.savefig('deep_complete_analysis.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
print("âœ“ Grafik 4 kaydedildi: deep_complete_analysis.png")
```
